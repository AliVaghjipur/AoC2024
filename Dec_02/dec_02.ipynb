```python
converted_invalid_dicts={}



for i in range(758):
    is_valid = False  # Flag to track if validity is determined

    for j in range(len(num_all[i])):
        # Create a temporary list excluding the current element
        temp_list = [x for x in num_all[i] if x != num_all[i][j]]

        # Check if the list is sorted (ascending or descending)
        if temp_list == sorted(temp_list) or temp_list == sorted(temp_list, reverse=True):
            # Calculate differences
            diff_list = np.diff(temp_list)

            # Check if differences are within the valid range
            if any(np.abs(diff) > 3 or np.abs(diff) < 1 for diff in diff_list):
                converted_invalid_dicts[i] = "InValid"
            else:
                converted_invalid_dicts[i] = "Valid"

            is_valid = True  # Set flag to break outer loop
            break  # Break inner loop
    
    # Break outer loop if validity is determined
    if is_valid:
        continue
    else:
        # If no valid temp_list is found, consider it invalid
        converted_invalid_dicts[i] = "InValid"
```
