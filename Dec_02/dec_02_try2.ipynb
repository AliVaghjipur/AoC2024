```python
import pandas as pd
import numpy as np
import re
```

```python
lines = [line.replace('\n', '') for line in open('input_2.txt', 'r')]

num_all = {}
for i, line in enumerate(lines):
    num_all[i] = list(map(int, re.findall(r'\d+', line)))
    
num_all
```

## Part 1
### answer: 242

```python
def validate_list(nums):
    """
    Validates if a list meets the conditions:
    1. All numbers are increasing or decreasing.
    2. Adjacent numbers differ by at least 1 and at most 3.
    """
    if nums == sorted(nums) or nums == sorted(nums, reverse=True):  # Check increasing or decreasing
        diffs = [abs(nums[i] - nums[i + 1]) for i in range(len(nums) - 1)]
        return all(1 <= diff <= 3 for diff in diffs)  # Check difference condition
    return False
```

```python
valid_dict= {}
for i in range(len(num_all)):
    valid_dict[i] = validate_list(num_all[i])
    
```

```python
valid_dict
```

```python
sum(1 for v in valid_dict.values() if v == True)
```

```python
def remove_to_validate(nums):
    """
    Removes numbers from the list to satisfy validation conditions.
    
    Args:
        nums (list): The original list of numbers.

    Returns:
        list: A valid list that satisfies the conditions.
    """
    if validate_list(nums):
        return nums  # Return the list if it already meets the conditions

    # Try removing each number and recursively validate
    for i in range(len(nums)):
        temp_list = nums[:i] + nums[i + 1:]  # Remove the current number
        if validate_list(temp_list):  # If removing the number creates a valid list
            return remove_to_validate(temp_list)  # Recursively check the reduced list
    
    return nums  # Return the list if no single removal makes it valid
```

```python
new_valid_dict = {}
for i in range(len(num_all)):
    new_valid_dict[i] = remove_to_validate(num_all[i])
```

```python
new_valid_dict
```

```python
new_dict= {}
for k in range(len(new_valid_dict)):
    new_dict[k] = validate_list(new_valid_dict[k])
```

```python
sum(1 for v in new_dict.values() if v == True)
```
